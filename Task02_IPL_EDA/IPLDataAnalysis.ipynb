```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (10,5)
```

```python
#loading dataset
df = pd.read_csv('IPL.csv',low_memory=False)
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df.columns=df.columns.str.strip().str.lower()
df.columns
```

```python
#Handling missing values
df.isnull().sum()
```

```python
#replacing missing values
df['player_of_match']=df['player_of_match'].fillna('Unknown')
df['venue']=df['venue'].fillna('Unknown')
df['match_won_by']=df['match_won_by'].fillna('No Result')
df['city']=df['city'].fillna('Unknown')
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
#ball level to match level
match_df=df.drop_duplicates(subset='match_id')
```

```python
#Exploratory data analysis
season_matches = match_df['season'].value_counts().sort_index()

sns.barplot(x=season_matches.index, y=season_matches.values,palette='deep',hue=season_matches.index,legend=False)
plt.title('Matches Played per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```

```python
team_wins = match_df['match_won_by'].value_counts().head(10)

sns.barplot(x=team_wins.values, y=team_wins.index,palette='coolwarm',hue=team_wins.index,legend=False)
plt.title('Top 10 Most Successful IPL Teams')
plt.xlabel('Wins')
plt.ylabel('Team')
plt.show()
```

```python
sns.countplot(x='toss_decision', data=match_df,color='teal')
plt.title('Toss Decision Distribution')
plt.show()
```

```python
#TOSS WINNER VS MATCH WINNER
match_df.loc[:,'toss_match_win'] = match_df['toss_winner'] == match_df['match_won_by']
(match_df['toss_match_win'].mean()) * 100
```

```python
venue_counts = match_df['venue'].value_counts().head(10)


sns.barplot(x=venue_counts.values, y=venue_counts.index,palette='pastel',hue=venue_counts.index,legend=False)
plt.title('Top 10 IPL Venues')
plt.xlabel('Matches Played')
plt.ylabel('Venue')
plt.show()
```

```python
top_batters = df.groupby('batter')['runs_batter'].sum().sort_values(ascending=False).head(10)


sns.barplot(x=top_batters.values, y=top_batters.index,palette='Set2',hue=top_batters.index,legend=False)
plt.title('Top 10 Run Scorers in IPL')
plt.xlabel('Runs')
plt.ylabel('Batter')
plt.show()
```

```python
top_bowlers = df['bowler'].value_counts().head(10)


sns.barplot(x=top_bowlers.values, y=top_bowlers.index,palette='viridis',hue=top_bowlers.index,legend=False)
plt.title('Top 10 Wicket Takers in IPL')
plt.xlabel('Wickets')
plt.ylabel('Bowler')
plt.show()
```

```python
over_runs = df.groupby('over')['runs_total'].mean()


sns.lineplot(x=over_runs.index, y=over_runs.values, marker='o',color='crimson',linewidth=2)
plt.title('Average Runs per Over')
plt.xlabel('Over')
plt.ylabel('Average Runs')
plt.show()
```

```python
pivot_table = df.pivot_table(values='runs_total', index='over', columns='innings', aggfunc='mean')


sns.heatmap(pivot_table, cmap='YlOrRd',annot=False)
plt.title('Average Runs Heatmap (Over vs Innings)')
plt.show()
```
